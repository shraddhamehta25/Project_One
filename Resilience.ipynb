```python
# Importing dependencies
import requests
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
from scipy.stats import linregress 

# Read in the dataframes for each state
resilience_texas = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_Texas.csv")
resilience_north_carolina = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_North_Carolina1.csv")
resilience_new_york = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_New_York1.csv")
```

```python
# Load data from CSV file
resilience_texas = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_Texas.csv")

# Calculate the CRSI statics for Texas
# Min, Max, Mean, Standard Deviation, Variance for Texas

min_texas = resilience_texas['CRSI'].min()
max_texas = resilience_texas['CRSI'].max()
mean_texas = resilience_texas['CRSI'].mean()
stdev_texas = resilience_texas['CRSI'].std()
variance_texas = resilience_texas['CRSI'].var()

# Print statistics result
print("Statistics for texas:")
print("Minimum CRSI:", min_texas)
print("Maximum CRSI:", max_texas)
print("Mean CRSI:", mean_texas)
print("Standard Deviation CRSI:", stdev_texas)
print("Variance CRSI:", variance_texas)
```

```python
# Load data from CSV file
resilience_north_carolina = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_North_Carolina1.csv")

# Calculate the CRSI statics for North Carolina
# Min, Max, Mean, Standard Deviation, Variance for North Carolina

min_north_carolina = resilience_north_carolina['CRSI'].min()
max_north_carolina = resilience_north_carolina['CRSI'].max()
mean_north_carolina = resilience_north_carolina['CRSI'].mean()
stdev_north_carolina = resilience_north_carolina['CRSI'].std()
variance_north_carolina = resilience_north_carolina['CRSI'].var()

# Print statistics result
print("Statistics for North Carolina:")
print("Minimum CRSI:", min_north_carolina)
print("Maximum CRSI:", max_north_carolina)
print("Mean CRSI:", mean_north_carolina)
print("Standard Deviation CRSI:", stdev_north_carolina)
print("Variance CRSI:", variance_north_carolina)


```

```python
# Load data from CSV file
resilience_new_york = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_New_York1.csv")

# Calculate the CRSI statics for New York
# Min, Max, Mean, Standard Deviation, Variance for New York

min_new_york = resilience_new_york['CRSI'].min()
max_new_york = resilience_new_york['CRSI'].max()
mean_new_york = resilience_new_york['CRSI'].mean()
stdev_new_york = resilience_new_york['CRSI'].std()
variance_new_york = resilience_new_york['CRSI'].var()

# Print statistics result
print("Statistics for New York:")
print("Minimum CRSI:", min_new_york)
print("Maximum CRSI:", max_new_york)
print("Mean CRSI:", mean_new_york)
print("Standard Deviation CRSI:", stdev_new_york)
print("Variance CRSI:", variance_new_york)

```

```python
# Perform an ANOVA test to assess the statistical significance differences between the states.

# Read in the dataframes for each state
resilience_texas = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_Texas.csv")
resilience_north_carolina = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_North_Carolina1.csv")
resilience_new_york = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_New_York1.csv")

# Combine the dataframes into one dataframe named 'data'
data = pd.concat([resilience_texas, resilience_north_carolina, resilience_new_york])

# Perform ANOVA test to compare the data from the three states
f_statistic, p_value = f_oneway(data[data['State'] == 'Texas']['CRSI'],
                                data[data['State'] == 'North Carolina']['CRSI'],
                                data[data['State'] == 'New York']['CRSI'])

# Print the results
print("F-Statistic:", f_statistic)
print("P-Value:", p_value)

# Interpret the results based on the p-value
if p_value < 0.05:
    print("The differences between the states are statistically significant.")
else:
    print("There is not enough evidence to conclude that the states have different resilience scores.")

```

```python
# ANOVA Visualization

# Read the data
resilience_texas = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_Texas.csv")
resilience_north_carolina = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_North_Carolina1.csv")
resilience_new_york = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_New_York1.csv")

# Combine the dataframes into one dataframe
data = pd.concat([resilience_texas, resilience_north_carolina, resilience_new_york])

# Perform ANOVA test
f_statistic, p_value = f_oneway(data[data['State'] == 'Texas']['CRSI'],
                                data[data['State'] == 'North Carolina']['CRSI'],
                                data[data['State'] == 'New York']['CRSI'])

# Print the ANOVA results
print("F-Statistic:", f_statistic)
print("P-Value:", p_value)

# Create a swarm plot
plt.figure(figsize=(8, 6))
for state in ['Texas', 'North Carolina', 'New York']:
    plt.scatter([state]*len(data[data['State'] == state]['CRSI']), data[data['State'] == state]['CRSI'], label=state)
plt.title('CRSI Distribution Across States')
plt.xlabel('State')
plt.ylabel('CRSI')
plt.legend()
plt.show()
```

```python
# Establish Visualizations for state Climate Resilience Index (CRSI)

# Create the bar chart and capture the bars
# Data
mean_values = [mean_texas, mean_north_carolina, mean_new_york]
states = ['Texas', 'North Carolina', 'New York']

# Create the bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(states, mean_values)

# Add data labels to each bar
for bar, value in zip(bars, mean_values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05, round(value, 2), ha='center', va='bottom')

plt.xlabel('State')
plt.ylabel('Mean CRSI')
plt.title('Mean Climate Resilience Screening Index for Each State')
plt.show()
```

```python
# Box plot for CRSI distribution

plt.figure(figsize=(8, 6))
plt.boxplot([resilience_texas['CRSI'], resilience_north_carolina['CRSI'], resilience_new_york['CRSI']], labels=states)
plt.xlabel('State')
plt.ylabel('CRSI')
plt.title('Climate Resilience Screening Index Distribution by State')
plt.show()
```

```python
# ANOVA plot for CRSI analysis
plt.figure(figsize=(8, 6))
plt.bar(['F-Statistic', 'P-Value'], [f_statistic, p_value])
plt.xlabel('ANOVA Results')
plt.ylabel('Value')
plt.title('ANOVA Results for CRSI Comparison')
plt.show()
```

```python
#Linear Regression (Texas):  Variables are CRSI and Risk

# Load your data
data = pd.read_csv("Resilience Resource/Climate_Resilience_Screening_Index_Texas.csv")

# Filter the data for Texas
texas_data = data[data['State'] == 'Texas']

# Create a scatter plot
plt.scatter(texas_data['CRSI'], texas_data['Risk'])
plt.xlabel('CRSI')
plt.ylabel('Risk')
plt.title('Texas: CRSI vs. Risk')

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(texas_data['CRSI'], texas_data['Risk'])

# Plot the regression line
plt.plot(texas_data['CRSI'], slope * texas_data['CRSI'] + intercept, color='red')

# Determine correlation and statistical significance
print("Correlation and Statistical Significance:")
print("R-value:", r_value)
print("P-value:", p_value)
print("Standard error:", std_err)
```
